```r
### print out the maxium value and minimum value of those three series
print(paste0("The maximum value of series1 is ", max(D1$x_values)))
print(paste0('the minimum value of series1 is ', min(D1$x_values)))
cat('\n')
print(paste0("The maximum value of series2 is ", max(D2$x_values)))
print(paste0('the minimum value of series2 is ', min(D2$x_values)))
cat('\n')
print(paste0("The maximum value of series3 is ", max(D3$x_values)))
print(paste0('the minimum value of series3 is ', min(D3$x_values)))
```

## 1.1 Plot the series

```r
par(mfrow=c(3,1))    # set the plotting area into a 3*1 array
plot(x=D1$num, y=D1$x_values, type="l", ylim=c(-7, 7), col=996, main="series plots", sub="series1", xlab="numbers", ylab="x_values")
plot(x=D2$num, y=D2$x_values, type="l", ylim=c(-7, 7), col=339, sub="series2", xlab="numbers", ylab="x_values")
plot(x=D3$num, y=D3$x_values, type="l", ylim=c(-7, 7), col=666, sub="series3", xlab="numbers", ylab="x_values")
```

## 1.2 Plot the acf functions for each series

```r
series1_acf = acf(x=D1$x_values, lag.max=15, type="correlation", plot=F)
series2_acf = acf(x=D2$x_values, lag.max=15, type="correlation", plot=F)
series3_acf = acf(x=D3$x_values, lag.max=15, type="correlation", plot=F)
```

```r
par(mfrow=c(3, 1)) 
plot(series1_acf, col=996, main="acf figures", sub="acf figures of series1")
plot(series2_acf, col=339, main="", sub="acf figures of series2")
plot(series3_acf, col=666, main="", sub="acf figures of series3")
```

## 1.2 Plot the pacf functions for each series

```r
series1_pacf = pacf(x=D1$x_values, lag.max=15, plot=F)
series2_pacf = pacf(x=D2$x_values, lag.max=15, plot=F)
series3_pacf = pacf(x=D3$x_values, lag.max=15, plot=F)
```

```r
par(mfrow=c(3, 1)) 
plot(series1_pacf, col=996, main="pacf figures", sub="pacf figures of series1")
plot(series2_pacf, col=339, main="", sub="pacf figures of series2")
plot(series3_pacf, col=666, main="", sub="pacf figures of series3")
```
